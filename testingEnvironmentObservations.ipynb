```python
import retro
from utils import *
import numpy as np
from gym.wrappers import FrameStack
from wrappers import *

env = retro.make(game="SonicTheHedgehog-Genesis", state="LabyrinthZone.Act1")
env.reset()

next_state, reward, done, info = env.step(POSSIBLE_ACTIONS[np.random.randint(0, len(POSSIBLE_ACTIONS))])
print(f"{next_state.shape},\n {reward},\n {done},\n {info}")
```

```python
# Apply Wrappers to environment
env = SkipFrame(env, skip=4)
env = GrayScaleObservation(env)
env = ResizeObservation(env, shape=(156, 156))
env = FrameStack(env, num_stack=4)

env.reset()
```

```python
import matplotlib.pyplot as plt
def plot_sequence_observations(nex_state):
    fig=plt.figure(figsize=(28, 15))

    for i in range(len(next_state)):
        ax = fig.add_subplot(1, 4, i+1)
        ax.imshow(next_state[i]) #Converting to matplotlib format
        ax.set_title(f"Frame {i+1}")
```

```python
next_state, reward, done, info = env.step(POSSIBLE_ACTIONS[np.random.randint(0, len(POSSIBLE_ACTIONS))])
print(f"{next_state.shape},\n {reward},\n {done},\n {info}")
plot_sequence_observations(next_state)
```
